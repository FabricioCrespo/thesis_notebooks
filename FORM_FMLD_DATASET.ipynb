11. Read Wider_testing.txt file and copy all the images into FMLD_dataset/test

```python
!mv /content/FMLD/Wider_dataset/widerface/val /content/FMLD/Wider_dataset/widerface/WIDER_val
```

```python
import os
import shutil

#Open the file and take all the lines (each line is an images)
with open("Wider_testing.txt") as f:
  lines=f.readlines()

print(lines)

path="/content/FMLD/Wider_dataset/widerface/"
dest="/content/FMLD/FMLD_dataset/test/"


for i in range(len(lines)):
  source=path+str(lines[i])
  source=source[:-1]
  shutil.copy2(source, dest)
  
```
