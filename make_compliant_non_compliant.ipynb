# Code to make compliant and no compliant datasets with crops from original images

## Import libraries

```python
import cv2
from xml.dom import minidom
import os
```

### Read all XML files for test

```python
rootPath="/content/drive/MyDrive/TESIS/Datasets/FMLD_dataset/FMLD_annotations/test/"
xml_test_files=os.listdir(rootPath)

rootImagePath="/content/drive/MyDrive/TESIS/Datasets/FMLD_dataset/test/"

maskedPath="/content/drive/MyDrive/TESIS/Datasets/FMLD_dataset/definitivo/test/Compliant/"
unmasked_facePath="/content/drive/MyDrive/TESIS/Datasets/FMLD_dataset/definitivo/test/Non-compliant/"



#Loop thought the folder
for xml in xml_test_files:
  mydoc = minidom.parse(rootPath+xml)

  #Crops Coordinates
  xMin=items = mydoc.getElementsByTagName('xmin')
  yMin=items = mydoc.getElementsByTagName('ymin')
  xMax=items = mydoc.getElementsByTagName('xmax')
  yMax=items = mydoc.getElementsByTagName('ymax')

  #Get image label

  labels=mydoc.getElementsByTagName('name')

  #Get image name
  image_name=mydoc.getElementsByTagName('filename')
  image_name=image_name[0].firstChild.data

  #Read image 
  img = cv2.imread(rootImagePath+image_name)

  #Make the crop
  for i in range(len(xMin)):
    x1=int(float(xMin[i].firstChild.data))
    x2=int(float(xMax[i].firstChild.data))
    y1=int(float(yMin[i].firstChild.data))
    y2=int(float(yMax[i].firstChild.data))

    #Make sure that min image size is greater or equal to 40px x 40px
    if (x2-x1)>=20 and (y2-y1)>=20:

      crop_img = img[y1:y2, x1:x2]
      if labels[i].firstChild.data=="masked_face":
        try:
          cv2.imwrite(maskedPath+image_name+"_"+str(i)+".jpg", crop_img)
          print(maskedPath+image_name.split(".")[0]+"_"+str(i)+".jpg")
        except:
            continue
      elif labels[i].firstChild.data=="unmasked_face":
        try:
          cv2.imwrite(unmasked_facePath+image_name.split(".")[0]+"_"+str(i)+".jpg", crop_img)
          print(unmasked_facePath+image_name.split(".")[0]+"_"+str(i)+".jpg")
        except:
          continue
      elif labels[i].firstChild.data=="incorrectly_masked_face":
        try:
          cv2.imwrite(unmasked_facePath+image_name+"_"+str(i)+".jpg", crop_img)
          print(unmasked_facePath+image_name.split(".")[0]+"_"+str(i)+".jpg")
        except:
          continue
    else:
      print("Non min size")
      continue
    
```

Read all XML files for train

```python
rootPath="/content/drive/MyDrive/TESIS/Datasets/FMLD_dataset/FMLD_annotations/train/"
xml_test_files=os.listdir(rootPath)

rootImagePath="/content/drive/MyDrive/TESIS/Datasets/FMLD_dataset/train/"

maskedPath="/content/drive/MyDrive/TESIS/Datasets/FMLD_dataset/definitivo/train/Compliant/"
unmasked_facePath="/content/drive/MyDrive/TESIS/Datasets/FMLD_dataset/definitivo/train/Non-compliant/"


cont=0
#Loop thought the folder
for xml in xml_test_files:
  mydoc = minidom.parse(rootPath+xml)

  #Crops Coordinates
  xMin=items = mydoc.getElementsByTagName('xmin')
  yMin=items = mydoc.getElementsByTagName('ymin')
  xMax=items = mydoc.getElementsByTagName('xmax')
  yMax=items = mydoc.getElementsByTagName('ymax')

  #Get image label

  labels=mydoc.getElementsByTagName('name')

  #Get image name
  image_name=mydoc.getElementsByTagName('filename')
  image_name=image_name[0].firstChild.data

  #Read image 
  img = cv2.imread(rootImagePath+image_name)

  #Make the crop
  for i in range(len(xMin)):
    x1=int(float(xMin[i].firstChild.data))
    x2=int(float(xMax[i].firstChild.data))
    y1=int(float(yMin[i].firstChild.data))
    y2=int(float(yMax[i].firstChild.data))

    #Make sure that min image size is greater or equal to 0px x 0px (it is not needed because there are only 31k faces and not 50 as the repo says)
    if (x2-x1)>=0 and (y2-y1)>=0:
      cont+=1
      crop_img = img[y1:y2, x1:x2]
      if labels[i].firstChild.data=="masked_face":
        try:
          cv2.imwrite(maskedPath+image_name+"_"+str(i)+".jpg", crop_img)
          print(maskedPath+image_name.split(".")[0]+"_"+str(i)+".jpg")
        except:
            continue
      elif labels[i].firstChild.data=="unmasked_face":
        try:
          cv2.imwrite(unmasked_facePath+image_name.split(".")[0]+"_"+str(i)+".jpg", crop_img)
          print(unmasked_facePath+image_name.split(".")[0]+"_"+str(i)+".jpg")
        except:
          continue
      elif labels[i].firstChild.data=="incorrectly_masked_face":
        try:
          cv2.imwrite(unmasked_facePath+image_name+"_"+str(i)+".jpg", crop_img)
          print(unmasked_facePath+image_name.split(".")[0]+"_"+str(i)+".jpg")
        except:
          continue 
      print(cont)
    else:
      print("Non min size")
      continue
print(cont)
```
