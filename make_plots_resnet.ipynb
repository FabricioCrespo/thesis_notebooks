## Accuracy on Test Dataset

```python
# Python program for implementation of Bubble Sort to Sort the values in the Graph
  
def bubbleSort(arr,txt_files):
    n = len(arr)
  
    # Traverse through all array elements
    for i in range(n-1):
    # range(n) also work but outer loop will repeat one time more than needed.
  
        # Last i elements are already in place
        for j in range(0, n-i-1):
  
            # traverse the array from 0 to n-i-1
            # Swap if the element found is greater
            # than the next element
            if arr[j] > arr[j+1] :
                arr[j], arr[j+1] = arr[j+1], arr[j]
                txt_files[j],txt_files[j+1]=txt_files[j+1], txt_files[j]
```

```python
import os
import random
import numpy as np


folder="/content/drive/MyDrive/TESIS/pesos_custom_dataset/15_epochs/"

files=os.listdir(folder)

txt_files=[]
accu_array=[]
labels=[]
for element in files:
  if (element.split(".")[-1])=="txt":
    txt_files.append(element)

for txt in txt_files:
  label_txt=txt.split("_")[0]
  labels.append(label_txt) 
  with open(folder+txt) as f:
      lines = f.readlines()
      for line in lines:
        if line.startswith("Test accuracy :"):
          accu_array.append(float(line.split(" ")[3])*100)
      #plt.plot(epochs, acc_train,  label=label_txt)

#print(accu_array)
#print(labels)
bubbleSort(accu_array,labels)

#print in reverse order, de mayor a menor xq o sino queda de menor a mayor 
print(accu_array[::-1])
print(labels[::-1])

import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

#vertical plot
#ax.bar(labels,accu_array)
#horizontal plot
colores=['black', 'red', 'green', 'blue', 'cyan','darkgoldenrod','pink','slategray','greenyellow','tomato']
#colores=np.random.shuffle(colores)

plt.barh(labels,accu_array,color=colores)
plt.xlabel('Accuracy %')
plt.ylabel('Model')
plt.title('Accuracy on Test Dataset')
plt.show()
```

```python
import os
import random
import numpy as np


folder="/content/drive/MyDrive/TESIS/pesos_custom_dataset/OptimizersVsLoss/"

files=os.listdir(folder)

txt_files=[]
accu_array=[]
labels=[]
for element in files:
  if (element.split(".")[-1])=="txt":
    txt_files.append(element)

for txt in txt_files:
  label_txt=txt.split("_")[0]
  optimizer=(txt.split("_")[-1]).split(".")[0]
  complete_name=label_txt+"_"+optimizer
  definitivo=complete_name.split("-")[0]+"-"+complete_name.split("-")[1]+"_"+(complete_name.split("_")[-1]).split(".")[0]
  labels.append(definitivo) 
  with open(folder+txt) as f:
      lines = f.readlines()
      for line in lines:
        if line.startswith("Test accuracy :"):
          accu_array.append(float(line.split(" ")[3])*100)
      #plt.plot(epochs, acc_train,  label=label_txt)

#print(accu_array)
#print(labels)
bubbleSort(accu_array,labels)

#print in reverse order, de mayor a menor xq o sino queda de menor a mayor 
print(accu_array[::-1])
print(labels[::-1])

import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

#vertical plot
#ax.bar(labels,accu_array)
#horizontal plot
colores=['black', 'red', 'green', 'blue', 'cyan','darkgoldenrod','pink','slategray','greenyellow','tomato']
#colores=np.random.shuffle(colores)

plt.barh(labels,accu_array,color=colores)
plt.xlabel('Accuracy %')
plt.ylabel('Model')
plt.title('Accuracy on Test Dataset')
plt.show()

```
